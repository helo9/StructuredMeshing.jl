```julia
using StructuredMeshing, Plots
```

```julia
meshdef = emptyMeshDef()

# create first vertex
v1 = addVertex(meshdef, [0.0, 0.0])
```

```julia
bound1 = extrude(meshdef, v1, [1.0, 0.0], 1.0, 10)
```

```julia
block1 = extrude(meshdef, bound1, [0.0, 1.0], 1.0, 10)

show(meshdef)
```
