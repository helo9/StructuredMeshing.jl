```julia
using Revise
using StructuredMeshing
```

```julia
meshdef = emptyMeshDef()

v1 = addVertex!(meshdef, [0.0, 0.0])

b1 = extrude!(meshdef, v1, [1.0, 0.0], 1.0, 11)

extrude!(meshdef, b1, [0.0, -1.0], 1.0, 9)

show(meshdef)
```

```julia
mesh1 = mesh(meshdef)

show(mesh1)
```

```julia
writeabq("test.inp", mesh1)
```

*test.inp* can be imported as part into Abaqus CAE
